```python
import sqlite3
import pandas as pd

from pathlib import Path

database_path = "Resources/ev_data_2012.sqlite"
#Path(database_path).touch()

conn = sqlite3.connect(database_path)
c = conn.cursor()


schema = '''CREATE TABLE IF NOT EXISTS 'ev_data_2012' ('fuel_type_code' TEXT,
   'station_name' TEXT,
   city TEXT,
   state TEXT,
   'status_code' TEXT,
   latitude FLOAT,
   longitude FLOAT,
   'groups_with_access_code' TEXT,
   index INT)'''

# c.execute(schema)

csv_2014_ev_data = pd.read_csv("CleanedCSVs/ev_data_2012.csv")
csv_2014_ev_data.to_sql("ev_data_2012", conn, if_exists='replace', index=False, dtype={'index': 'PRIMARY KEY'})

    
c.execute("SELECT * FROM 'ev_data_2012'")
i = c.fetchall()
for row in i:
    print(row)

conn.close()
```
