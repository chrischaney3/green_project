```python
import sqlite3
import pandas as pd

from pathlib import Path

database_path = "sqlite/2021_ev_data.sqlite"
Path(database_path).touch()

conn = sqlite3.connect(database_path)
c = conn.cursor()

c.execute('''CREATE TABLE data_2021 (Fuel Type Code TEXT,
  Station Name TEXT,
  City TEXT,
  State TEXT,
  Status Code TEXT,
  Latitude TEXT,
  Longitude TEXT,
  Groups With Access Code TEXT)''')


csv_2021_ev_data = pd.read_csv("CSV_Data/2021_ev_data.csv")
csv_2021_ev_data.to_sql("2021_ev_data", conn, if_exists='append', index=False)

conn.close()
```
